# Machine Learning
- independent data is always on 2-D
- 100% accuracy = overfit
- -ve accuracy = underfit

## Linear Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
f = pd.read_csv("demo.csv")
df = pd.DataFrame(f)
x=df[["in"]] #independent data is always in 2-D
y=df["out"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=LinearRegression()
ans=model.fit(x_train,y_train)
pre=model.predict(x_test)
acc=r2_score(y_test,pre)
print(f"Accuracy is {acc}")
mse = mean_squared_error(y_test,pre)
mse
```

```python
i = model.intercept_
print(i)
```

```python
# Slope
s = model.coef_
print(s)
```

### Question

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
f = pd.read_csv("SATGPA.csv")
df = pd.DataFrame(f)
x=df[["SAT"]] #independent data is always in 2-D
y=df["GPA"]
l=[1,2,4,5,6]
r2=[]
m2=[]
for i in l:
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    model=LinearRegression()
    ans=model.fit(x_train,y_train)
    pre=model.predict(x_test)
    acc=r2_score(y_test,pre)
    r2.append(acc)
    
    mse = mean_squared_error(y_test,pre)
    m2.append(mse)
print(f"r2 score {r2}")
print(f"mse {m2}")
```
