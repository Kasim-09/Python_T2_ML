````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
````

````python
f = pd.read_csv("Fish.csv")
df = pd.DataFrame(f)
x = df.drop("Species",axis=1)
y = df["Species"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = KNeighborsClassifier(n_neighbors=4)
model.fit(x_train,y_train)
pre = model.predict(x_test)
acc = accuracy_score(y_test,pre)
acc
````

````python
f = pd.read_csv("Fish.csv")
df = pd.DataFrame(f)
z = []
for i in range(10):
    x = df.drop("Species",axis=1)
    y = df["Species"]
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    model = KNeighborsClassifier(n_neighbors=4)
    model.fit(x_train,y_train)
    pre = model.predict(x_test)
    acc = accuracy_score(y_test,pre)  
    z.append(acc)
print(z)
    
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
f = pd.read_csv("diabetes.csv")
df = pd.DataFrame(f)
x = df.drop("Outcome",axis=1)
y = df["Outcome"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
pre = model.predict(x_test)
acc = accuracy_score(y_test,pre)
print(acc)
z=confusion_matrix(y_test,pre)
print(z)

p = classification_report(y_test,pre)
print(p)
````

# RandomForestClassification


````python
from sklearn.ensemble import RandomForestClassifier
````

````python
f = pd.read_csv("drug200.csv")
df = pd.DataFrame(f)
x = df.drop(["Drug"],axis=1)
y = df["Drug"]
x = pd.get_dummies(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = RandomForestClassifier(n_estimators=50,criterion="entropy",max_depth=5,random_state=42)
model.fit(x_train,y_train)
pre = model.predict(x_test)
acc = accuracy_score(y_test,pre)
# z = confusion_matrix(y_test,pre)
# p = classification_report(y_test,pre)
print(acc)
````

````python
f = pd.read_csv("cancer.csv")
df = pd.DataFrame(f)
x = df.drop(["id","diagnosis","Unnamed: 32"],axis=1)
y = df["diagnosis"]
x = pd.get_dummies(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = RandomForestClassifier(n_estimators=5,criterion="entropy",max_depth=3,random_state=42)
model.fit(x_train,y_train)
pre = model.predict(x_test)
acc = accuracy_score(y_test,pre)
z = confusion_matrix(y_test,pre)
p = classification_report(y_test,pre)
print(acc)
print(z)
print(p)
````
